```python
#SRI 12

alt.Chart(df).mark_rect().encode(
    x='Year:O',
    y='Month:O',
    color=alt.Color(
        'SRI12', scale=alt.Scale(scheme='redblue', domain=(-2, 2))),
    tooltip=[
        alt.Tooltip('Year:O', title='Year'),
        alt.Tooltip('Month:O', title='Month'),
        alt.Tooltip('mean(SRI12):Q', title='SRI12 Indix')
    ]).properties(width=600, height=250)
```

```python
alt.Chart(df).mark_bar(size=2).encode(
    x='Date:T',
    y='SRI12:Q',
    color=alt.Color(
        'SRI12:Q', scale=alt.Scale(scheme='redblue', domain=(-2, 2))),
    tooltip=[
        alt.Tooltip('Date:T', title='Date'),
        alt.Tooltip('SRI12:Q', title='SRI12')
    ]).properties(width=600, height=250)
```
